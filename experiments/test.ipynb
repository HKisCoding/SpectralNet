```python
import pandas as pd
```

### Colon cancer

```python
df_patient = pd.read_csv('../dataset/colon_cancer/Colorectal Cancer Patient Data.csv', index_col= 0) 
df_patient.info()
```

```python
df_patient.head()
```

```python
df_gene = pd.read_csv('../dataset/colon_cancer/Colorectal Cancer Gene Expression Data.csv', index_col= 0)
df_gene.info()
```

```python
df_gene = df_gene.transpose()
# let's check the result:
df_gene.head()
```

```python
df_combined = df_patient.join(other=df_gene, on='ID_REF', how='left')
df_combined
```

```python
df_combined.to_csv('../dataset/colon_cancer/colon_cancer.csv')
```
